```python
import pandas as pd
import matplotlib.pyplot as plt
```

Importing Dataset

```python
df = pd.read_csv("C:/Users/theja/Desktop/DataJobs/Glassdoor_Job_Postings.csv")
df
```

Drop null Values

```python
df.dropna(inplace=True)
df
```

Removing Duplicates

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace = True)
df
```

Reseting Index

```python
df.reset_index(inplace = True)
df
```

```python
rows, columns = df.shape
print(f"The data has {rows} rows and {columns} columns.")
```

```python
unique_employment_types = df['employment_type'].unique()
print(unique_employment_types)
```

```python
unique_sectors = df['sector'].unique()
print(unique_sectors)
```

```python
df.info()
```

```python
data_scientist_jobs = df[df['job_title'].str.contains('Data Scientist|Data Science',case= False, na = False)]
data_scientist_jobs['job_title'].head(20)
```

```python
len(data_scientist_jobs)
```

```python
data_analyst_jobs =  df[df['job_title'].str.contains('Data Analyst|Data Analytics|Analytics', case = False, na=False)]
data_analyst_jobs['job_title'].head(20)
```

```python
len(data_analyst_jobs)
```

```python
ds_employment_type_counts = data_scientist_jobs['employment_type'].value_counts()
plt.figure(figsize=(10,6))
ds_employment_type_counts.plot(kind='bar',color='skyblue')
plt.title('Distribution of company types for Data Scientist jobs')
plt.xlabel('Company Type')
plt.ylabel('Number of jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
ds_employment_type_counts = data_analyst_jobs['employment_type'].value_counts()

plt.figure(figsize=(10,6))
ds_employment_type_counts.plot(kind='bar',color='lightgreen')
plt.title('Distribution of company types for Data Analytics jobs')
plt.xlabel('Company Type')
plt.ylabel('Number of jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
ds_valid_sectors = data_scientist_jobs[data_scientist_jobs['sector']!='--']
ds_sectors = ds_valid_sectors['sector'].value_counts().head(5)

plt.figure(figsize=(10,6))
ds_sectors.plot(kind='bar',color='skyblue')
plt.title('Distrbution of sector for Data Scientist Jobs')
plt.xlabel('Sector')
plt.ylabel('Number of jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
ds_valid_sectors = data_analyst_jobs[data_analyst_jobs['sector']!='--']
ds_sectors = ds_valid_sectors['sector'].value_counts().head(5)

plt.figure(figsize=(10,6))
ds_sectors.plot(kind='bar',color='lightgreen')
plt.title('Distrbution of sector for Data Analyst Jobs')
plt.xlabel('Sector')
plt.ylabel('Number of jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
ds_averages = data_scientist_jobs[['career_opportunities_rating', 'comp_and_benefits_rating','culture_and_values_rating','senior_management_rating','work_life_balance_rating']].mean()
da_averages = data_analyst_jobs[['career_opportunities_rating', 'comp_and_benefits_rating','culture_and_values_rating','senior_management_rating','work_life_balance_rating']].mean()

ratings_comparision = pd.DataFrame({'Data Scientist': ds_averages,'Data Analyst': da_averages})
ratings_comparision.plot(kind='bar',figsize=(10,6))

plt.title('Comparision of rating between Data Scientists and Data Analysts Jobs')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--',alpha=0.7)
plt.show()
```
